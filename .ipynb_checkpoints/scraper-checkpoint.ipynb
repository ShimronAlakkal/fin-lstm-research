```python
from bs4 import BeautifulSoup as bs
import requests as req
import os 
```

```python
res_americas = req.get('https://www.google.com/finance/markets/indexes/americas')
res_me = req.get('https://www.google.com/finance/markets/indexes/europe-middle-east-africa')
res_asia = req.get('https://www.google.com/finance/markets/indexes/asia-pacific')
html_americas = bs(res_americas.text, 'html.parser')
html_me = bs(res_me.text, 'html.parser')
html_asia = bs(res_asia.text, 'html.parser')
```

```python
index_names_class = 'ZvmM7'
index_price_class = 'YMlKec'
index_change_class = 'P2Luy Ebnabc'
index_cperc_class = 'JwB6zf'
americas_index_names = html_americas.find_all("div",class_ = index_names_class)
americas_index_price = html_americas.find_all("div",class_ = index_price_class)
americas_index_change = html_americas.find_all("span",class_ = index_change_class)
americas_index_cperc = html_americas.find_all("span",class_ = index_change_cperc)
for i in range(len(americas_index_change)):
    print(americas_index_names[i].get_text(), americas_index_price[i].get_text(), americas_index_change[i].get_text(), americas_index_cperc[i].get_text(),'\n')
```

```python
index_names_class = 'ZvmM7'
me_index_names = html_me.find_all("div",class_ = index_names_class)
for i in me_index_names:
    print(i.get_text(),'\n')
```

```python
index_names_class = 'ZvmM7'
ap_index_names = html_asia.find_all("div",class_ = index_names_class)
for i in ap_index_names:
    print(i.get_text(),'\n')
```
