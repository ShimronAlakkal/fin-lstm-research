```python
from bs4 import BeautifulSoup as bs
import requests as req
import os 
```

## The global market indices

```python
americas_index_url = 'https://www.google.com/finance/markets/indexes/americas'
me_index_url = 'https://www.google.com/finance/markets/indexes/europe-middle-east-africa'
ap_index_url = 'https://www.google.com/finance/markets/indexes/asia-pacific'
```

```python
res_americas = req.get(americas_index_url)
res_me = req.get(me_index_url)
res_ap = req.get(ap_index_url)

html_americas = bs(res_americas.text, 'html.parser')
html_me = bs(res_me.text, 'html.parser')
html_ap = bs(res_ap.text, 'html.parser')
```

```python
index_names_class = 'ZvmM7'
index_price_class = 'YMlKec'
index_change_class = 'P2Luy Ebnabc'
index_cperc_class = 'JwB6zf'
```

```python
americas_index_names = html_americas.find_all("div",class_ = index_names_class)
americas_index_price = html_americas.find_all("div",class_ = index_price_class)
americas_index_change = html_americas.find_all("span",class_ = index_change_class)
americas_index_cperc = html_americas.find_all("div",class_ = index_cperc_class)

me_index_names = html_me.find_all("div",class_ = index_names_class)
me_index_price = html_me.find_all("div",class_ = index_price_class)
me_index_change = html_me.find_all("span",class_ = index_change_class)
me_index_cperc = html_me.find_all("div",class_ = index_cperc_class)

ap_index_names = html_ap.find_all("div",class_ = index_names_class)
ap_index_price = html_ap.find_all("div",class_ = index_price_class)
ap_index_change = html_ap.find_all("span",class_ = index_change_class)
ap_index_cperc = html_ap.find_all("div",class_ = index_cperc_class)
```

```python
index_data_map = {}
temp_americas, temp_me, temp_ap = {}, {}, {}

for i in range(len(americas_index_change)):
    temp_americas[americas_index_names[i].get_text()] = {'ltp': americas_index_price[i].get_text(), 'change' : americas_index_change[i].get_text(), 'perc change' : americas_index_cperc[i].get_text()}
index_data_map['americas'] = temp_americas

for i in range(len(me_index_change)):
    temp_me[me_index_names[i].get_text()] = { 'ltp': me_index_price[i].get_text(), 'change': me_index_change[i].get_text(), 'perc change' : me_index_cperc[i].get_text()}
index_data_map['me'] = temp_me

for i in range(len(ap_index_change)):
    temp_ap[ap_index_names[i].get_text()] = {'ltp': ap_index_price[i].get_text(), 'change': ap_index_change[i].get_text(), 'perc change': ap_index_cperc[i].get_text()}
index_data_map['ap'] = temp_ap
```

## Top Ganiers and Top Losers
